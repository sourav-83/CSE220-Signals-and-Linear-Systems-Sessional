```python
class Student:
    def __init__(self, name, batch, roll):
        self.name = name
        self.batch = batch
        self.roll = roll
    def __str__(self):
        return f"Name: {self.name}\nBatch: {self.batch}\nRoll: {self.roll}"
```

```python
class player:
    def __init__(self, name, club):
        self.name = name
        self.club = club
    def __str__(self):
        return f"Name: {self.name}\nClub: {self.club}"
```

```python
messi = player("messi", "barca")

print(messi)
```

```python
student = Student("Messi", 22, 10)
print(student)
```

```python
pred = np.array([8,3,8,3])
label = np.array([2,2,2,2])

mse = (1/len(pred)) * np.sum(np.square(pred-label))
mse
```

```python
cross_entropy_loss = (-1/len(pred)) * np.sum(np.multiply(label, np.log(pred)))
cross_entropy_loss
```
