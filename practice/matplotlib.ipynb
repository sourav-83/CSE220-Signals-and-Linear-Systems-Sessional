```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)

plt.figure(figsize=(4,4))
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label = 'quadratic')
plt.plot(x, x ** (-1), label = 'inverse')
plt.xlabel('x')
plt.ylabel('f (x)')
plt.title('simple plot')
plt.legend()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)
plt.figure(figsize=(3,3))
plt.plot(x,label='x')
plt.plot(x**2+3, label = 'f(x)')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
x = np.linspace(0, 10 * np.pi, 10000)
y = np.sin(x)
plt.figure()
plt.plot(x,y)
```

```python
plt.scatter(x,y,linewidths=.001)
```

```python
plt.bar(x,y)
```

```python
fig, axs = plt.subplots(2,2, figsize=(8,8))
axs[0][0].plot(x,y)
axs[0][1].scatter(x,y)
axs[1][0].bar(x,y)
axs[1][1].boxplot(x)
fig.show()
```
