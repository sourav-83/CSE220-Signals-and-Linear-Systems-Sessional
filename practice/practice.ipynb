```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(-10, 11)

x = np.where((n >= 0) & (n <= 4), 1, 0)

print(x)
```

```python
plt.style.use('ggplot')
```

```python
plt.stem(n, x)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
plt.show()
```

**Time Shifting** 

$y[n] = x[n - k]$

```python
k = 3
n_shifted = n - k  

plt.stem(n_shifted, x)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
plt.show()
```

```python
k = 3
x_shifted = np.roll(x,-k)

plt.stem(n, x_shifted)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
plt.show()
```

**Time Reversal**

$y[n] = x[-n]$

```python
x_reversed = np.flip(x)
n_reversed = -np.flip(n) 

plt.stem(n_reversed, x_reversed)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
```

**Odd and Even Decomposition**

$x_e[n] = \frac{x[n] + x[-n]}{2}$

$x_o[n] = \frac{x[n] - x[-n]}{2}$

```python
x_rev = np.flip(x)
even = np.multiply(.5, np.add(x,x_rev))
odd = np.multiply(.5, np.subtract(x,x_rev))

plt.stem(n, np.add(even,odd))
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
```

**Time Scaling (Downsampling)**

$y[n] = x[Mn]$

```python
M = 3
x_down = x[::M]
n_down = n[::M]

plt.stem(n_down, x_down)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
```

**Time Scaling (Upsampling)**

$y[n] = x[n/L]$

```python
L = 2

x_up = np.zeros(len(x) * L)

x_up[::L] = x 

n_up = np.arange(-10 * L, (10*L)+2)


plt.stem(n_up, x_up)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
```

```python
x_interpolated = np.interp(n_up, n_up[::L], x)


plt.stem(n_up, x_interpolated)
plt.xlabel('n (Time Index)')
plt.ylabel('x[n] (Amplitude)')
plt.title('Discrete Time Signal')
plt.grid(True)
```
